```python
import cv2 
from matplotlib import pyplot as plt

image = cv2.imread("/media/a22nicolesdc/DIURNOEXTRA/programacion/detector-caras-py/docs/igmgm.jpg")
imaxe_color = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
gray = cv2.cvtColor(imaxe_color, cv2.COLOR_BGR2GRAY)

plt.imshow(imaxe_color)
plt.title('Imaxe1')
plt.show()
plt.imshow(gray)
plt.title('Imaxe en gris')
plt.show()
```

```python
def buscarCaras(imaxe):
    ruta_xml = cv2.data.haarcascades + "haarcascade_frontalface_default.xml"
    face_cascade = cv2.CascadeClassifier(ruta_xml)

    if len(imaxe.shape) == 3:
        imaxe_gris = cv2.cvtColor(imaxe, cv2.COLOR_RGB2GRAY)
    else:
        imaxe_gris = imaxe

    # esots parametros ahyq ue r ajustandolos hasta que mas omenos
    faces = face_cascade.detectMultiScale(
        imaxe_gris,
        scaleFactor = 1.2,
        minNeighbors = 10,
        minSize = (100, 100)
    )
    
    print(f"Caras encontradas: {len(faces)}")
    
    destino = imaxe.copy()
    
    color_rosa = (180, 105, 255) 
    grosor = 16
    
    for (x,y,w,h) in faces:
        cv2.rectangle(destino, (x,y), (x+w, y+h), color_rosa, grosor)

    plt.figure(figsize=(10,6))
    plt.imshow(destino)
    plt.title(f'Detectadas: {len(faces)}')
    plt.axis('off')
    plt.show()

imageColor = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
buscarCaras(imageColor)
```
